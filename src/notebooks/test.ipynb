```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression

print("NumPy:", np.__version__)
print("Pandas:", pd.__version__)

# Tiny toy dataset
X = [[0, 0], [1, 1], [2, 2], [-1, -1]]
y = [0, 1, 1, 0]

model = LogisticRegression().fit(X, y)

print("Predict [0, 0] ->", model.predict([[0, 0]])[0])
print("Predict [1, 1] ->", model.predict([[1, 1]])[0])
print("Predict [2, 2] ->", model.predict([[2, 2]])[0])
print("Predict [-1, -1] ->", model.predict([[-1, -1]])[0])
```
